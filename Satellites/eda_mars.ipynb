Clean IMEO MARS


Dowload csv with no input details required or via api

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
download_url = 'https://methanedata.unep.org/map?mars=&company=&country=&cc=&sat=&limit=1000&pub=#mcoord=0/30/0'
# download_date = '03122025'
download_date = '11122025'
directory     = 'satellite/data/mars' 
fname_in      = f'unep_methanedata_detected_plumes_{download_date}.csv'
fname_out     = f'unep_methanedata_detected_plumes_cleaned_{download_date}.csv'
```

```python
#----------------------------------------------------------------------------------------------------------
# sensor, country, date, emission rate (kg/h), uncertainty (kg/h), latitude, longitude, comments, platform | 
#----------------------------------------------------------------------------------------------------------
```

```python
#mars = pd.read_csv(os.path.join(cmm_dir, 'mars', fname_in))
mars = pd.read_csv(os.path.join(directory, fname_in))

mars.columns = mars.columns.str.strip() 

mars.rename(columns={'satellite': 'sensor'}, inplace=True)
mars.rename(columns={'tile_date': 'date'}, inplace=True)
mars.rename(columns={'lat': 'latitude'}, inplace=True)
mars.rename(columns={'lon': 'longitude'}, inplace=True)
mars.rename(columns={'ch4_fluxrate': 'emission rate (kg/h)'}, inplace=True)
mars.rename(columns={'ch4_fluxrate_std': 'uncertainty (kg/h)'}, inplace=True)

# deal with wierd date fmt 2023-02-24T04:11:58
mars['date'] = pd.to_datetime(mars['date'], format='mixed')
mars['date'] = mars['date'].dt.strftime('%d/%m/%Y')

mars['country'] = mars['country'].replace('United States', 'USA')
mars['country'] = mars['country'].replace('United States of America', 'USA')

mars['comments'] =  f'Downloaded from {download_url} on the {download_date}. Cleaned using eda_mars.ipynb'
mars['platform'] = 'IMEO MARS'

columns_to_keep = ['sensor', 'sector', 'country','date', 'emission rate (kg/h)', 'uncertainty (kg/h)', 'latitude', 'longitude', 'comments', 'platform']
mars = mars[columns_to_keep]

mars.to_csv(os.path.join(directory, fname_out), index=False) 

print(mars.columns)
print(mars.shape)
print(mars['sensor'].unique())
```
