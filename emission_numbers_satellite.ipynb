```python
import pandas as pd
```

```python
plumes = pd.read_csv(
    r'C:\Users\rebek\Ember\plume_perc\Plumes.csv', # New GEM plume dataset
    thousands=",",       # interpret commas as thousand separators inside numbers
    quotechar='"',       # respect quoted fields
    engine="python"      # more flexible parser for tricky CSVs
)

total_emissions = (
    plumes.groupby("Country/Area")["Emissions (kg/hr)"]
      .sum()
      .reset_index(name="TOTAL_EMISSIONS_kg_per_hr")
)

total_emissions
```
