```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    r'C:\Users\rebek\Downloads\query_result_2025-11-17T12_05_30.75753068Z.csv',
    thousands=",",       # interpret commas as thousand separators inside numbers
    quotechar='"',       # respect quoted fields
    engine="python"      # more flexible parser for tricky CSVs
)

df
```

```python
# Select and rename relevant columns
df_clean = df[["YEAR", "PRODUCTION_MT", "SOURCE"]].rename(
    columns={
        "YEAR": "year",
        "PRODUCTION_MT": "production",
        "SOURCE": "source"
    }
)

# Drop rows without emissions data
df_clean = df_clean.dropna(subset=["production"])

# Global totals per year and source
global_by_year_source = (
    df_clean.groupby(["year", "source"], as_index=False)["production"].sum()
)

# Save results
global_by_year_source.to_csv("global_production_by_year_source.csv", index=False)
```
