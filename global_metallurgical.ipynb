```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    r'C:\Users\rebek\Ember\Data Tool\Data_for_flourish\cmm-data-tool\eia_production.csv', #File outputted from eia_met_production.ipynb
    thousands=",",       # interpret commas as thousand separators inside numbers
    quotechar='"',       # respect quoted fields
    engine="python"      # more flexible parser for tricky CSVs
)

df
```

```python
# Select and rename relevant columns
df_clean = df[["year", "metallurgical coal (MT)"]].rename(
    columns={
        "metallurgical coal (MT)": "production"
    }
)

# Drop rows without emissions data
df_clean = df_clean.dropna(subset=["production"])

# Global totals per year and source
global_by_year_source = (
    df_clean.groupby(["year"], as_index=False)["production"].sum()
)

# Save results
global_by_year_source.to_csv("global_met_by_year_source.csv", index=False)
```
