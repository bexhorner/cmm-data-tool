```python
import pandas as pd
import pycountry
```

```python
sat = pd.read_csv(
    r'C:\Users\rebek\Ember\Data Tool\cmm-data-tool\merged_kayrros_mars_carbonmapper_02102025_coal.csv', # New GEM plume dataset
    thousands=",",       # interpret commas as thousand separators inside numbers
    quotechar='"',       # respect quoted fields
    engine="python"      # more flexible parser for tricky CSVs
)
```

```python
plumes_per_country = (
    sat
    .groupby("country")
    .size()
    .reset_index(name="plume_count")
    .sort_values("plume_count", ascending=False)
    .reset_index(drop=True)
)

country_map = {
    "USA": "United States of America",
    "Russian Federation": "Russia"
}

sat["country_clean"] = sat["country"].replace(country_map)

plumes_per_country = (
    sat
    .groupby("country_clean")
    .size()
    .reset_index(name="plume_count")
    .sort_values("plume_count", ascending=False)
    .reset_index(drop=True)
)

plumes_per_country
```

```python
plumes_per_country.to_csv("plume_counts.csv", index=False)
```
