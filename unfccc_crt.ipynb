```python
import unfccc_di_api
import numpy as np
import pandas as pd
```

```python
def get_detailed_by_party(country_code):
    reader = unfccc_di_api.ZenodoReader()
    df = reader.query(party_code=country_code)
    df_ch4_1b1a = df[(df["gas"] == "CH4") & (df["category"] == '1.B.1.a  Coal Mining and Handling')]
    df_ch4_1b1a = df_ch4_1b1a[df_ch4_1b1a['year'] != 'Base year']
    df_ch4_1b1a = df_ch4_1b1a[df_ch4_1b1a['measure'] == 'Net emissions/removals']
    return df_ch4_1b1a
```

```python
annex_i_countries = [
    "AUS", "AUT", "BEL", "BGR", "CAN", "HRV", "CYP", "CZE", "DNK", "EST",
    "FIN", "FRA", "DEU", "GRC", "HUN", "ISL", "IRL", "ISR", "ITA", "JPN",
    "LVA", "LIE", "LTU", "LUX", "MLT", "NLD", "NZL", "NOR", "POL", "PRT",
    "ROU", "RUS", "SVK", "SVN", "ESP", "SWE", "CHE", "TUR", "GBR", "USA"
]

all_data = []
for i in annex_i_countries:
    df = get_detailed_by_party(i)
    all_data.append(df)

all_df = pd.concat(all_data)

all_df
```
